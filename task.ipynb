```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def people(documents):
    doc_num = input("Введите номер документа: ")
    for i in documents:
        if i['number'] == doc_num:
            print('Владелец документа: ',i['name'])
            break
    else:
        print('Документ в базе не найден')

def shelf(directories):
    doc_num = input("Введите номер документа: ")
    for i in directories:
        if doc_num in directories[i]:
            print('Номер полки: ',{i})
            break
    else:
        print('Документ в базе не найден')

def all_info(documents):
    for i in documents:
        type = i['type']
        number = i['number']
        name = i['name']
        for j in directories:
            if number in directories[j]:
                print(f"№ документа: {i['number']};", f"Тип документа: {i['type']};", f"Владелец: {i['name']};", f"Номер полки: {j}")


def add_shelf(directories):
    number = input("Введите номер полки : ")
    if number in directories:
        print('Такая полка уже существует. Текущий перечень полок:' + ','.join(directories))
    else:
        directories[number] = []
        print('Полка добавлена. Текущий перечень полок:' + ','.join(directories))
        
        
def del_shelf(directories):
    number = input("Введите номер полки : ")
    if number not in directories:
        print('Такой полки не существует. Текущий перечень полок:' + ', '.join(directories))
    elif len(directories[number]) == 0:
        del directories[number]
        print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories))
    else:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories))
            
def ad(directories,documents):
    doc_type = input("Введите тип документа: ")
    doc_number = input("Введите номер документа: ")
    doc_name = input("Введите имя владельца документа: ")
    shelf_id = input("Введит номер полки: ")
    if shelf_id not in directories:
        print('Полки не существует')
    new_doc = dict(type=doc_type, number=doc_number, name=doc_name)
    documents.append(new_doc)
    directories[shelf_id] += [doc_number]
    print('Документ успешно добавлен')
    

def main(user_input):
    while True:
        if user_input == 'p':
            people(documents)
        elif user_input == 's':
            shelf(directories)
        elif user_input == 'l':
            all_info(documents)
            break
        elif user_input == 'as':
            add_shelf(directories)
        elif user_input == 'ds':
            del_shelf(directories)
        elif user_input == 'ad':
            ad(directories,documents)
            break
        elif user_input == 'm':
            m(directories)
        elif user_input == 'q':
            print('Конец работы')
            break
        else:
            print('Вы ввели некорректную команду, повторите ввод.')
            break


main(user_input = input('Введите команду: '))
```
